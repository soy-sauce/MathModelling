```python
import numpy as np
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
```

```python
dataset_url='http://tuvalu.santafe.edu/~aaronc/powerlaws/data/words.txt'
data = pd.read_csv(dataset_url, sep=" ", header=None)
```

```python
from scipy.stats import itemfreq
tmp = itemfreq(data) # Get the item frequencies
x = tmp[:, 0] # unique values in data
y = tmp[:, 1] # freq

import matplotlib.pyplot as plt
plt.loglog(x, y, basex=2, basey=2, linestyle='None', marker='o')
#plt.plot(x,y,"o")
plt.show()
```

```python
coefficients = np.polyfit(np.log10(x), np.log10(y), 1)
polynomial = np.poly1d(coefficients)
log10_y_fit = polynomial(np.log10(x))  

plt.plot(x, y, 'o')
plt.plot(x, 10**log10_y_fit, '*')     
plt.yscale('log')
plt.xscale('log')
print(polynomial)
#s=-0.9338
```
