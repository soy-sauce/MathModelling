```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
import scipy.signal
```

```python
def f(X, t, sigma, b, r):
    x, y, z = X  
    return sigma * (y - x), x * (r - z) - y, x * y - b * z  
```

```python
r = 28
sigma = 10
b  = 8/3

X0 = [1.0, 1.0, 1.0]  # initial vector X0 = [x(0), y(0), z(0)]

t = np.arange(0.0, 40.0, 0.01)

X = odeint(f, X0, t, args=(sigma,b,r))

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(t, X[:,2])
ax.grid(True)
ax.set(xlabel='t')
ax.legend(['z'],loc="lower right")
```

```python
r = 28
sigma = 10
b  = 8/3

X0 = [1.0, 1.0, 1.0]  # initial vector X0 = [x(0), y(0), z(0)]

t = np.arange(0.0, 100.0, 0.1)

X = odeint(f, X0, t, args=(sigma,b,r))

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(t, X[:,2])
ax.grid(True)
ax.set(xlabel='t')
ax.legend(['z'],loc="lower right")
```

```python
peaks=scipy.signal.find_peaks(X[:,2])
print(peaks[0])
print(len(peaks[0]))
peaks=peaks[0]
```

```python
print(peaks)
zn=peaks[:-1]
znplus1=peaks[1:]
print(zn)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(znplus1,zn)
ax.grid(True)
ax.set(xlabel='z_n+1')
ax.set(ylabel='z_n')
```

If we draw a cobweb plot we can see there are no periodic orbits since the function increases at a constant rate that is almost the same rate as y=x.
